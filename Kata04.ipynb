## Ejercicio 1: Transformar cadenas



```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""
```

Dividir el texto en cada oración para trabajar con su contenido

```python
texto=text.split('. ')
texto
```

Define algunas palabras clave para búsqueda que te ayudarán a determinar si una oración contiene un hecho.

```python
palabras=["average", "temperature", "distancie"]
```

Crear un bucle para imprimir solo datos sobre la Luna que estén relacionados con las palabras clave definidas anteriormente:

```python
for n in texto:
    for palabra in palabras:
        if palabra in n:
              print(n)
```

Actualiza el bucle(ciclo) para cambiar C a Celsius

```python
for n in texto:
    for palabra in palabras:
        if palabra in n:
              print(n.replace('C', 'Celsius'))
```

Crear una plantilla de cadena multilínea para contener el resto de la información. En lugar de usar kilómetros, debes convertir la distancia a metros multiplicando por 1,000.

```python
facts= f"{'-'*80} \nName= {name}\nGravity in {name} = {round(gravity*1000,3)} m/s2\nPlanet= {planet}"

print(facts)
```

Finalmente, usa ambas variables para unir el título y los hechos.

```python
plantilla=f"{title.title()}\n{facts}"
print(plantilla)
```
