```python
values_moon=planet_moons.values()
planets=len(planet_moons)
```

```python
total_moons= 0
for i in values_moon:
 total_moons= total_moons+i
print(f"""Hay {total_moons} lunas en total""")
print(f"""Hay {total_moons/planets} lunas en promedio por planeta""")
```
